This is the notebook that we'll start with for day 2 of the Python tutorial.  It should have all of the code that we need from day 1.  A difference is that this notebook now uses the standard abbreviations for NumPy and Matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt
```

In the following file, rows correspond to individual patients and columns correspond to daily inflammation measurements for that patient.

```python
data = np.loadtxt(fname='data/inflammation-01.csv', delimiter=',')
```

We use the following code to inspect statisical properties of `data`.

```python
fig = plt.figure(figsize=(10.0, 3.0))

axes1 = fig.add_subplot(1, 3, 1)
axes2 = fig.add_subplot(1, 3, 2)
axes3 = fig.add_subplot(1, 3, 3)

axes1.set_ylabel('average')
axes1.plot(np.mean(data, axis=0))

axes2.set_ylabel('max')
axes2.plot(np.max(data, axis=0))

axes3.set_ylabel('min')
axes3.plot(np.min(data, axis=0))

fig.tight_layout() 

plt.show()
```
